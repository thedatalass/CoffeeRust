```python
#CMD - pip install pytz, pip install pandas, pip install matplotlib, pip install numpy
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
#Read the CSV into a data frame
df = pd.read_csv('Coffee-Rust.csv')
df
```

```python
#Exploratory data analysis
#Let's start by looking at the summary statistics for each variable in the data.
df.describe()

#Rain ranges from 0.2-407.7 mm/month, Temperature ranges from 23.36-27.16C, Rust ranges from 0.33-50%,
#Production ranges from 80.33-3832.67 1000-60kg bags/month and futures range from 21.98-175.18 US$.
```

```python
#Let's look at a plot of all columns with labels
df.plot();
plt.title("Exploring Coffee Rust, Production and Futures")
```

```python
#Since the regular scale plot isn't very helpful, let's look at the variable interaction on a log scale.
df.plot();
plt.yscale('log')
plt.title("Log - Coffee Rust, Production and Futures")

#This is obviously not the best visualization to see what's happening between the variables.
```

```python
#Now, let's look at a plot of only the Futures variable over time. 
df['Futures'].plot(color="#ac7339")
plt.ylabel("Futures in USD",fontsize=16)
plt.title("Coffee Futures over Time",fontsize=18)
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
#From this plot we can see that the Futures prices fluctuate quite a bit.
#Also note the straight line at about 135 and 250 on the y-axis. These would be the data gaps in the original data so
#using this visualization would be misleading for the research.

#Recall original data is from 1998-1991, 1995, 2005-2006, 2008-2009 and 2011-2013.
```

```python
#Now, let's look at the relationship between Production and Futures.

df['Production'].plot(color='black')
df['Futures'].plot(color="#ac7339")

plt.xlabel("Futures Prices in US$",fontsize=16)
plt.ylabel("Production in 1000-60 kg bags",fontsize=16)
plt.title("Coffee Production versus Futures",fontsize=18)
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
plt.legend()

#From this plot, we see that there is a much greater range in the 'Production' variable than the 'Futures' variable.
#'Production' ranges from 80-3832.67 1000-60 kg bags produced per month. From this visualization, it's hard to see
# any possible relationship between production and futures variables
```

```python
#Next, let's look at relationship between Rust and Production
df['Rust'].plot(color='#663300')
df['Production'].plot(color='black')
plt.xlabel("Percent Rust on Plant",fontsize=16)
plt.ylabel("Production in 1000-60 kg bags",fontsize=16)
plt.title("Coffee Rust versus Production",fontsize=18)
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
plt.legend()

#From this plot, it's hard to tell if there's a correlation between rust and production variables.
```

```python
#Finally, let's take a look at the relationship between Rust and Futures to find any insights from the visualization.
#Note, we change the x-axis maximum to 50 to reflect maximum rust %.
df['Rust'].plot(color='black')
df['Futures'].plot(color='#ac7339')
#Set x-axis range from 0-50
plt.xlim(0,50)
#Add xaxis label
plt.xlabel("Percent Rust on Plant", fontsize=16)
#Add yaxis label
plt.ylabel("Futures in USD", fontsize=16)
plt.title("Coffee Rust versus Futures", fontsize=18)
plt.legend()
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
#There seems to be positive correlation between the percent rust on the plant and futures. 
```

```python
#Next, Let's look at a histogram to see how the Futures Price data is distributed.
df['Futures'].hist(bins=30, normed=True, color='#ac7339')
plt.xlabel("Coffee Futures in USD",fontsize=16)
plt.ylabel("Percent Distribution",fontsize=16)
plt.title("Coffee Futures Histogram",fontsize=18)
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
#Coffee futures range from 20-180 with the majority of futures occuring between 100-150 US$.
#A bin size of 30 seems to show the variations and groupings of the data the best.
```

```python
df['Rust'].hist(bins=20, normed=True, color='#ac7339')
plt.xlabel("Percent Coffee Rust on Leaves",fontsize=16)
plt.ylabel("Percent Distribution",fontsize=16)
plt.title("Coffee Rust Histogram",fontsize=18)
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)

#The Coffee rust ranges from 0-50% with the majority occurring between 5-20%.
#This is the best bin size to show variances in data.
```

```python
#Finally, let's look at Production variable
df['Production'].hist(bins=10, normed=True, color='#ac7339')
plt.xlabel("Production in 1000-60 kg bags", fontsize=16)
plt.ylabel("Percent Distribution",fontsize=16)
plt.title("Coffee Production Histogram",fontsize=18)

plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
#Coffee production ranges from 0-4000 l000-60 kg bags with most production occurring at 1000, 
#and 3000-4000 1000-60 kg bags.

#This visualization really doesn't reveal any patterns.
```

```python
#In addition to using histograms and kernel density estimates, let's use a Heatmap data visualization 
#to see relationships between the variables in the coffee data.

import seaborn as sns
ax = plt.axes()
corr = df.corr()
#The default colors were changed to 'BrBG_r' to match better with the 
#browns in the other visualizations.
sns.heatmap(corr, cmap='BrBG_r',xticklabels=corr.columns.values, yticklabels=corr.columns.values)
#Label title
ax.set_title('Correlation Matrix of Variables',fontsize=18)
plt.show()
#Rain and temperature are not correlated to each other.
#Rain is negatively correlated (about -0.6) with production, meaning that if the rain decreases, production increases.
#Temperature is positively correlated with production so if temp increases, so does production.
#Rust and production are slightly positively correlated so if rust increases, production increases slightly.
#Production and futures are positively correlated so if production increases, futures increase.
#Rust and futures are negetively correlated with each other so if rust increases, futures decrease.
```

```python
#Next, we could use Seaborn library for regression visualization of Rust versus Production
ax = plt.axes()
ax.set_title('Linear Regression of Percent Rust and Production',fontsize=16)
ax.set_xlabel('Percent Rust on Plant',fontsize=16)
ax.set_ylabel('Production in 1000-60kg bags',fontsize=16)
sns.regplot(x=df['Rust'],color='#ac7339', y=df['Production'],scatter_kws={"color":"black","alpha":0.3,"lw":4} )
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
sns.plt.show()
```

```python
#Regression visualization of Production versus Futures
ax = plt.axes()
ax.set_title('Linear Regression of Production versus Futures',fontsize=16)
ax.set_xlabel('Production in 1000-60kg bags',fontsize=16)
ax.set_ylabel('Futures in USD', fontsize=16)
sns.regplot(x=df['Production'], color='black', y=df['Futures'],line_kws={"color":"#ac7339","alpha":0.3,"lw":5} )
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
sns.plt.show()
```

```python
#regression visualization of Rust versus Futures
import seaborn as sns
ax = plt.axes()
ax.set_title('Linear Regression of Percent Rust vs. Futures',fontsize=16)
ax.set_xlabel('Percent Rust on Plant', fontsize=16)
ax.set_ylabel('Futures in USD', fontsize=16)
sns.regplot(x=df['Rust'], color='black', y=df['Futures'],line_kws={"color":"#ac7339","alpha":0.3,"lw":5} )
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
sns.plt.show()
#There's something unusual about this linear regression so let's use a lowess smoother.
#This approach has the fewest assumptions.
```

```python
#Let's use a pivot table to group the data by Rust percent on plant.
pivot_df = df.pivot_table(index=['Rust'])
pivot_df
```

```python
#Save df to CSV
pivot_df.to_csv('pivot.csv')
```

```python
#Let's see error and slope of each variable
import numpy as np
coefficients, residuals, _, _, _ = np.polyfit(range(len(pivot_df.index)),pivot_df,1,full=True)
mse = residuals[0]/(len(pivot_df.index))
nrmse = np.sqrt(mse)/(pivot_df.max() - pivot_df.min())

#Print results
print('Slope ' + str(coefficients[0]))
print('NRMSE: ' + str(nrmse))
```

```python
#Now let's manipulate the dataframe to only look at 'Rust' and 'Futures' variables.
df2 = pd.read_csv('rust_futuresonly.csv')
df2
```

```python
#Out of curiosity, let's do a correlation matrix of 'df2'.

import seaborn as sns
ax = plt.axes()
corr = df2.corr()
#The default colors were changed to 'BrBG_r' to match better with the 
#browns in the other visualizations.
sns.heatmap(corr, cmap='BrBG_r',xticklabels=corr.columns.values, yticklabels=corr.columns.values)
#Label title
ax.set_title('Correlation Matrix of Rust and Futures',fontsize=18)
plt.show()

#This shows that rust and futures are positively correlated to one another, which is the opposite of what
#the original correlation matrix from 'df' shows above. Should we fail to accept this original hypothesis based on this
#new evidence?
```

```python
#use least squares polynomial fit for Rust & Futures variables since linear regression didn't work well.
import numpy as np
coefficients, residuals, _, _, _ = np.polyfit(range(len(df2.index)),df2,1,full=True)
mse = residuals[0]/(len(df2.index))
nrmse = np.sqrt(mse)/(df2.max() - df2.min())

#Print results
print('Slope ' + str(coefficients[0]))
print('NRMSE: ' + str(nrmse))

#The slope is positive as shown in the correlation matrix above with minimum error so the polynomial regression
#fit is the more accurate visualization that should be used to make conclusions about the hypothesis.
```
