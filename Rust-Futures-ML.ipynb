```python

#CMD - pip install pytz
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline

#Can historical rust data predict futures prices?
```

```python
#Read the CSV into a data frame
df = pd.read_csv('Coffee-Rust.csv')
df
```

```python
#Since the Date, Country, Rain & Temp variables are not correlated to rust and/or futures, we'll delete these from the 
#dataframe.
coffee = df.drop(df.columns[[0, 1, 2, 3]], axis=1)
coffee.head()
```

```python
#Next, let's see if we can use "Rust" and "Production" variables to predict "Futures"
#X feature variable is Rust which is column 0 in the dataframe
X = df.values[:,0]
#Y target variable is Futures, which is column 2 in the dataframe.
y = df.values[:,2]
```

```python
#Use column 3 (Futures) as the feature and columns 0 (Rust) as the training variables
y = coffee.iloc[:, 2:3]
X = coffee.iloc[:, 0:1]
```

```python
#make sure the shape of the data is the same for X and y.
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#Decision Tree Regressor since data is continuous
#Import libraries
from sklearn.tree import DecisionTreeRegressor

#Create regressor model object
regressor = DecisionTreeRegressor(random_state=0)

#Fit the model
regressor.fit(X,y)
```

```python
#Great tutorial: http://www.ritchieng.com/machine-learning-evaluate-classification-model/
#make class predictions for the testing set
y_pred_class = regressor.predict(X_test)
```

```python
# calculate accuracy
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred_class))
```

```python
print(metrics.confusion_matrix(y_test, y_pred_class))
```
